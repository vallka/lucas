```python
forecast = {}
for p in places:
    r=requests.get(f'https://api.openweathermap.org/data/2.5/onecall?lat={p[1]}&lon={p[2]}&exclude=hourly,alert,current,minutely&units=metric&appid={key}')
    data = json.loads(r.text)
    print(p[0], data)
    forecast[p[0]] = {}
    forecast[p[0]]['daily'] = data['daily']
```
